```python
from rs import Coefficients, Strain, ResidualStress
```

```python
coef = Coefficients()

E = 70e3 # MPa
nu = 0.3
hole = 1.8 # mm

coef.diameter_correction(hole)
strain = Strain('Data/al.txt')
stress = ResidualStress(strain=strain, E=E, nu=nu, coef=coef, smoothing=True)
stress.smoothing = True
```

```python
strain.principal_strain(interpolation='poly', k=6)
```

```python
stress.show_stresses('normal')
```

```python
stress.tikhonov_correction()
```

```python
stress.show_stresses('normal')
```

```python
stress.show_stresses('principal')
```

```python
stress.report()
```
